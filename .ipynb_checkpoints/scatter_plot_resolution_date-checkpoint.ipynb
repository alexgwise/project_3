```python
import pandas as pd
import numpy as np
import panel as pn
import plotly.express as px
from pathlib import Path
pn.extension("plotly")
```

```python
# Read in the Dataset
file_path = Path("Resources/reimbursement_data.csv")
reimburse_df = pd.read_csv(file_path)
reimburse_df.head()
```

```python
scatter_plot = px.scatter(
    reimburse_df,
    x="Days to resolve",
    y="Adjustment Amount",
    color ="Resolution",
    title ="Adjustment Amounts Over Time",
)
```

```python
pane = pn.pane.Plotly(scatter_plot)
```

```python
pane
```

```python
write_off_df = reimburse_df.loc[reimburse_df["Resolution"]=="Write Off"]
```

```python
write_off_df.head()

bar_plot = px.bar(
    write_off_df,
    x="Manufacturer Name",
    y="Adjustment Amount",
    color ="Manufacturer Rejection Reason",
    title ="Adjustment Amounts Over Time",
)
```

```python
pane2 = pn.pane.Plotly(bar_plot)
```
