```python
# imports
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Read in data
reim = pd.read_csv(
    Path("../Project 3/reimbursement_data.csv")
    )

reim.head()
```

```python
avg_adjustment_by_id = reim.groupby("Contract ID").mean().sort_values(by='Adjustment Amount', ascending=False).head(10)
avg_adjustment_by_id
```

```python
top_ten_adj_id = avg_adjustment_by_id.hvplot.bar(x='Contract ID', y='Adjustment Amount', rot=90, xlabel="Contract ID", ylabel="Adjustment")
top_ten_adj_id
```

```python
order_to_invoice = reim.groupby("Contract ID").sum().sort_values(by='days from order to invoice')
order_to_invoice
```

```python
days_order_invoice = order_to_invoice.hvplot(x='Contract ID', y='days from order to invoice', kind='scatter')
days_order_invoice
```

```python
scat = reim.hvplot.scatter(
               x="Adjustment Amount",
               y="Reimbursement Amount",
               by="Days to resolve",
               width=700, height=400,
               size=70, alpha=5,
               xlim=(0.0, 10000),
               ylim=(0.0, 10000),
               xlabel="Adjustment Amount",
               ylabel="Reimbursement Amount",
               title="Adjustment v. Reimbursement")

scat
```
